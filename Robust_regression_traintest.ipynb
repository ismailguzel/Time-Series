```r
library(MASS)
```

```r
data <- Boston
num_rows <- nrow(data) # ncol(data) number of column of dataset

set.seed(100)  # set seed to reproduce data

training_rows <- sample(1:num_rows, 0.8*num_rows)  # row indices of training data

training_data <- data[training_rows, ] # 404 rows  # make training data

test_data <- data[-training_rows, ] # 102 rows  # make test data
```

```r
lmMod <- lm (medv ~ crim + zn + chas + nox + rm + dis + ptratio +black +lstat, data)  # make lm model

robMod <- rlm (medv ~ crim + zn + chas + nox + rm + dis + ptratio +black +lstat, data)  # make robust model
```

```r
summary(lmMod)
```

```r
summary(robMod)
```

```r
lm_Predicted <- predict(lmMod, test_data)

rob_Predicted <- predict(robMod, test_data)
```

```r
lm_actuals_pred <- cbind(lm_Predicted, test_data$medv)

rob_actuals_pred <- cbind(rob_Predicted, test_data$medv)

mean(apply(lm_actuals_pred, 1, min)/ apply(lm_actuals_pred, 1, max)) 

mean(apply(rob_actuals_pred, 1, min)/ apply(rob_actuals_pred, 1, max))
```

```r
cDist <- cooks.distance (lmMod) # get cooks distance

resids <- stdres (lmMod) #errors

allObs <- cbind (data, cDist, resids)

allObs[cDist > 4/num_rows, ] # show outlier observations 
```
