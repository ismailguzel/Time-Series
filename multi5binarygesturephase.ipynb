```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
data=pd.read_csv('a1_raw.csv')
print(data["phase"].value_counts())
data.head(4)
```

Son kolonda 5 type class var. Multiclass olan veriyi multi binary olarak yazıyorum. Her Binary için svm çalıştırcam.

```python
data=np.array(data)
new1=np.array(list(range(data.shape[0])))
new2=np.array(list(range(data.shape[0])))
new3=np.array(list(range(data.shape[0])))
new4=np.array(list(range(data.shape[0])))
new5=np.array(list(range(data.shape[0])))
for i in list(range(data.shape[0])):
    if data[i,19]=='Rest':
            new1[i]=1
    else:
            new1[i]=0
    if data[i,19]=='Stroke':
            new2[i]=1
    else:
            new2[i]=0
    if data[i,19]=='Retraction':
            new3[i]=1
    else:
            new3[i]=0
    if data[i,19]=='Preparation':
            new4[i]=1
    else:
            new4[i]=0
    if data[i,19]=='Hold':
            new5[i]=1
    else:
            new5[i]=0
```

```python
newdata=np.column_stack([data[:,:19], new1, new2, new3, new4, new5])
pd.DataFrame(newdata).head(2)
```

```python
X = np.array(newdata)[:,:19]
y1 = np.array(newdata)[:,19]
y2 = np.array(newdata)[:,20]
y3 = np.array(newdata)[:,21]
y4 = np.array(newdata)[:,22]
y5 = np.array(newdata)[:,23]
X_train, X_test, y1_train, y1_test, y2_train, y2_test, y3_train, y3_test, y4_train, y4_test, y5_train, y5_test = train_test_split(X, y1, y2, y3, y4, y5, test_size=0.75)
```

```python
clf = SVC(kernel='linear')
clf.fit(X_train, y1_train)
```
